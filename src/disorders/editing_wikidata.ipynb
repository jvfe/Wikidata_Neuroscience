Creating quickstatements for the reconciled neural disorders

```python
import pandas as pd
```

```python
disord = pd.read_csv("../../data/disorders/doids.csv")
phenos = pd.read_csv("../../data/disorders/phenos.csv")
```

```python
disord['doid'] = disord['doid'].str.replace("obo.", "").str.replace("_", ":")
phenos['hpid'] = phenos['hpid'].str.replace("obo.", "").str.replace("_", ":")
```
