In this notebook I acquire the human phenotype IDs for the neural disorders in the reconciled list.

```python
from owlready2 import *
import pandas as pd
```

```python
neuro = pd.read_csv('../../data/disorders/recon-neuro-disorders_list.csv')
phenos = neuro.copy(deep=True)
```

```python
hp = get_ontology("http://purl.obolibrary.org/obo/hp.owl").load()
```

```python
hpid = []
for lab in phenos['Disorder']:
    hpid.append(hp.search_one(label = lab))
phenos['hpid'] = [str(i) for i in hpid]
```

```python
ind = phenos['hpid'].str.contains('None')
phenos = phenos[~ind]
phenos
```

```python
phenos.to_csv("../../data/disorders/phenos.csv")
```
